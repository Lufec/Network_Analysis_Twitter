```python
NUM_TWEETS_TO_FETCH = 17900

search_tweets_all = []

dates = ['2022-02-11','2022-02-12','2022-02-13','2022-02-14','2022-02-15','2022-02-16','2022-02-17']

for i in range(7)
    cursor = twitter.cursor(twitter.search, q='Ukraine AND Russia', count=100, result_type='mixed', until = dates[i])
    search_tweets = list(itertools.islice(cursor, NUM_TWEETS_TO_FETCH))
    search_tweets_all.append(search_tweets)
    sleep(901)
```

```python
i = 0
for search_tw in search_tweets_all:
    retweets = []
    for tweet in search_tw:
        if 'retweeted_status' in tweet:
            retweets.append(tweet)
    dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': [], 'retweet_count':[],'retweeter_usr':[]}
    for tweet in retweets:
        dict_['user'].append(tweet['user']['screen_name'])
        dict_['date'].append(tweet['created_at'])
        dict_['text'].append(tweet['text'])
        dict_['favorite_count'].append(tweet['favorite_count'])
        dict_['retweet_count'].append(tweet['retweet_count'])
        retweeted_status = tweet['retweeted_status']
        dict_['retweeter_usr'] = retweeted_status['user']['screen_name']
        
    # Structure data in a pandas DataFrame for easier manipulation
    df = pd.DataFrame(dict_)
    df.to_csv('dados_'+str(i)+'.csv',index = False, sep = ';')
    i = i+1
```

```python
import glob

path = r'/home/lufec/Downloads/redes' # use your path
all_files = glob.glob(path + "/*.csv")
li = []

for filename in all_files:
    df = pd.read_csv(filename, sep = ';', lineterminator = '\n')
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
import networkx as nx

D = nx.DiGraph()

for index, row in frame.iterrows():
    retweeted_sn = row['user']
    retweeter_sn = row['retweeter_usr']
    # Edge direction: retweeted_sn -> retweeter_sn
    if D.has_edge(retweeter_sn, retweeted_sn):
        D.edges[retweeter_sn, retweeted_sn]['weight'] += 1
    else:
        D.add_edge(retweeter_sn, retweeted_sn, weight=1)
```

```python
max(D.nodes, key=D.out_degree)
```
